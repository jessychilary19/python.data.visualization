Mudança de Index 

```python
#executando no mesmo dataframe
df.set_index('País', inplace=True)
```

```python
# criando uma variavel (anos) para diminuir as colunas
anos = list(map(str,range(1980,2014)))
```

```python
# Trazer somente os dados de brasil
brasil= df.loc['Brasil',anos]
```

```python
brasil
```

```python
#renomeando as colunas
brasil_dict= {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
dados_brasil= pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

Criando os gráficos com Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

Mudando o tamanho o gráfico

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()    
```

Adcionando o titulo do gráfico e dos eixos x e y

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Qtd de Imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()    
```

Criando uma figura para customizar os gráficos

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel(' Qtd de Imigrantes')
plt.show()
```

Plotar 2 graficos numa mesma figura

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Qtd de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Qtd de Imigrantes')
axs[1].grid()

plt.show()
```

Analisando a tendencia de imigração para os países da America do Sul

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Imigração dos 4 maiores Países da América do Sul\n1980 a 2013')
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Qtd de Imigrantes')


ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)


plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index,america_sul['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Qtd de Imigrantes', fontsize= 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
# Ordenando do maior pro menor valor
america_sul_sorted= america_sul.sort_values('Total',ascending=True)
```

```python
cores = []
for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul: Brasil foi o 4° pais com mais imigrantes\npara o Canadá 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Qtd de Imigrantes', fontsize= 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Adicionando Rotulo de dados (totais)

```python
cores = []
for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul: Brasil foi o 4° pais com mais imigrantes\npara o Canadá 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Qtd de Imigrantes', fontsize= 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# adicionando rotulo de dados
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha= 'left', va='center')

#retirar a borda
ax.set_frame_on(False)

# retirar o eixo x
ax.get_xaxis().set_visible(False)

ax.tick_params(axis='both', which='both', length=0)


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Qtd de imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)

plt.show()
```

Alterando cores com paletas de cores

Documentação da biblioteca seaborn = https://seaborn.pydata.org/tutorial/color_palettes.html

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

```python
# Retirar as bordas do gráfico
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('QTD de Imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```
